# Popularity Based Recommender system


```python
ratings_with_name = ratings.merge(books,on='ISBN')
ratings_with_name.head()
```

```python
num_rating_df = ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating_df.rename(columns = {'Book-Rating':'num_ratings'},inplace=True)
num_rating_df
```

```python
avg_rating_df = ratings_with_name.groupby('Book-Title').mean()['Book-Rating'].reset_index()
avg_rating_df.rename(columns = {'Book-Rating':'avg_ratings'},inplace=True)
avg_rating_df
```

```python
popular_df = num_rating_df.merge(avg_rating_df,on="Book-Title")
popular_df
```

```python
popular_df = popular_df[popular_df['num_ratings']>=250].sort_values('avg_ratings',ascending=False).head()
```

```python
popular_df = popular_df.merge(books,on="Book-Title").drop_duplicates('Book-Title')[['Book-Title','Book-Author','Image-URL-M','num_ratings','avg_ratings']]
```

# Collaborative Filtering Based Recommender System

```python
x = ratings_with_name.groupby('User-ID').count()['Book-Rating'] > 200
read_write_users = x[x].index
```

```python
filtered_ratings = ratings_with_name[ratings_with_name['User-ID'] .isin(read_write_users)]
filtered_ratings.head()
```

```python
y = filtered_ratings.groupby('Book-Title').count()['Book-Rating']>=50
famous_books = y[y].index
```

```python
final_ratings = filtered_ratings[filtered_ratings['Book-Title'].isin(famous_books)]
```

```python
final_ratings.drop_duplicates()
```

```python
pivot_table = final_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pivot_table.fillna(0,inplace=True)
```

```python
pivot_table
```

```python
import sklearn
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score = cosine_similarity(pivot_table)
```

```python
similarity_score.shape
```

```python
def recommend(book_name):
    #user_input = request.form.get('user_input')
    index = np.where(pivot_table.index == book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_score[index])),key=lambda x:x[1],reverse=True)[1:5]
    data = []
    for i in similar_items:
        item = []
        temp_df = books[books['Book-Title'] == pivot_table.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))

        data.append(item)

    return data
```

```python
recommend('The Notebook')
```
